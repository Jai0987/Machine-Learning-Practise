```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv("Advertising.csv")
```

```python
df.head()
```

```python
df['TotalSpend'] = df['TV'] + df['radio'] + df['newspaper']
```

```python
df.head()
```

```python
sns.scatterplot(data=df, x='TotalSpend', y='sales')
# ScatterPlot
```

```python
sns.regplot(data=df, x='TotalSpend', y='sales')
# Regular Plot
```

```python
X = df['TotalSpend']
y = df['sales']
```

```python
# y = mx + b
# y = B1x + B0

help(np.polyfit)
```

```python
np.polyfit(X,y,deg=1)
```

```python
# (B1, B0) returned
```

```python
potentialSpend = np.linspace(0,500,100)
```

```python
predictedSales = 0.04868788*potentialSpend + 4.24302822
```

```python
sns.scatterplot(x="TotalSpend", y='sales', data=df)
plt.plot(potentialSpend, predictedSales, color='green')
```

```python
sns.scatterplot(x="TotalSpend", y='sales', data=df, color='red')
plt.plot(potSpend, predSales, color='purple')
```
