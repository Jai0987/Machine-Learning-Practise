```python
# Create X and y
X = df.drop('sales', axis=1)
y = df['sales']

# TRAIN TEST SPLIT
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3, random_state=101)

# SCALE DATA
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test) 
```

```python
from sklearn.model_selection import cross_validate
from sklearn.linear_model import Ridge
```

```python
model = Ridge(alpha=100)
```

```python
scores = cross_validate(model, X_train, y_train, scoring=['neg_mean_absolute_error', 'neg_mean_squared_error'],cv=10)
```

```python
from sklearn.linear_model import ElasticNet
```

```python
baseElasticNetModel = ElasticNet()
```

```python
paramGrid = {'alpha':[0.1,1,5,10,50,100], 'l1_ratio':[0.1,0.5,0.7,0.95,0.99,1]}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
gridModel = GridSearchCV(estimator=baseElasticNetModel, param_grid=paramGrid, 
                         scoring='neg_mean_squared_error', cv=5,verbose=2)
```

```python
gridModel.fit(X_train, y_train)
```

```python
gridModel.best_estimator_
```

```python
gridModel.best_params_
```

```python
pd.DataFrame(gridModel.cv_results_)
```

```python
yPred = gridModel.predict(X_test)
```
